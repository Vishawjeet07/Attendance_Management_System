```python
import os
```

```python
class Student:

    def __init__(self, sid, name, age, course):
        self.sid = sid
        self.name = name
        self.age = age
        self.course = course

    def save_student(self):
        file = open("students.txt", "a")
        file.write(self.sid + "," + self.name + "," + self.age + "," + self.course + "\n")
        file.close()
```

```python
class AttendanceSystem:

    def add_student(self):
        sid = input("Student ID : ")
        name = input("Name : ")
        age = input("Age : ")
        course = input("Course : ")

        s = Student(sid, name, age, course)
        s.save_student()

        print("Student Added Successfully")
```

```python
system = AttendanceSystem()
system.add_student()
```

```python
def view_students():

    if not os.path.exists("students.txt"):
        print("No Record")
        return

    file = open("students.txt","r")

    for line in file:
        data = line.strip().split(",")

        print("ID :",data[0])
        print("Name :",data[1])
        print("Age :",data[2])
        print("Course :",data[3])
        print("----------------")

    file.close()
```

```python
view_students()
```
